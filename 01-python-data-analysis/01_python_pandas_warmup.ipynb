```python
import pandas as pd

data = {
    "Student": ["Aylin", "Bora", "Cem", "Deniz", "Ece"],
    "Math": [85, 72, 90, 66, 78],
    "Python": [88, 75, 95, 70, 82],
    "SQL": [80, 68, 92, 60, 85]
}

df = pd.DataFrame(data)

df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Average"] = df[["Math", "Python", "SQL"]].mean(axis=1)

df
```

```python
high_performance = df[df["Average"] >= 80]

high_performance
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df["Student"], df["Average"])
plt.title("Average Scores by Student")
plt.xlabel("Student")
plt.ylabel("Average Score")
plt.show()
```
